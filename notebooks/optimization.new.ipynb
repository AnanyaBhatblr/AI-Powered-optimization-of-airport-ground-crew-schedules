```python
import pandas as pd
```

```python
crew_df = pd.read_csv("../data/crew_data.csv")  # File containing crew details
task_df = pd.read_csv("../data/task_personnel_summary.csv")  # File containing task requirements
```

```python
assignments = []  # List to store crew-task assignments
available_crews = crew_df["Crew_ID"].tolist()  # All crew members initially available
```

```python
for _, task in task_df.iterrows():
    task_id = task["task_ID"]
    floor = task["Floor_No"]
    gate = task["Gate_number"]
    personnel_required = task["Number_of_Personnel"]
```

```python
# Map task requirements: {(task_ID, Floor_No, Gate_number): Number_of_Personnel}
task_requirements = {
    (row.task_ID, row.Floor_No, row.Gate_number): row.Number_of_Personnel
    for row in task_df.itertuples()
}
```

```python
# Step 2: Iterate over tasks and assign crews
for _, task in task_df.iterrows():
    task_id = task["task_ID"]
    floor = task["Floor_No"]
    gate = task["Gate_number"]
    personnel_required = task["Number_of_Personnel"]
    
    # Step 3: Assign available crew members to the task
    assigned_crews = available_crews[:personnel_required]  # Take the required number of crews
    available_crews = available_crews[personnel_required:]  # Update available crews list
    
    # Store assignments
    for crew_id in assigned_crews:
        assignments.append({
            "Crew_ID": crew_id,
            "task_ID": task_id,
            "Floor_No": floor,
            "Gate_number": gate
        })
    
    # If not enough crews are available, print a warning (optional)
    if len(assigned_crews) < personnel_required:
        print(f"Warning: Not enough crews to fully staff task {task_id} at Floor {floor}, Gate {gate}")
```

```python
assignments_df = pd.DataFrame(assignments)
```

```python
print("\nTask Assignments:")
print(assignments_df)
assignments_df.to_csv("task_assignments.csv", index=False)
```
